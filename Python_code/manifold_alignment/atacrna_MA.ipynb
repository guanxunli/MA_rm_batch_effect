```python
import numpy as np
import pandas as pd
import scanpy as sc
from MA_utility import *
```

```python
# load data
rna_path = "~/Documents/Single cell/MA_rm_batch_effect/data set/dna_atac/dta_rna.csv"
atac_path = "~/Documents/Single cell/MA_rm_batch_effect/data set/dna_atac/dta_atac.csv"
dta_rna = pd.read_csv(rna_path, index_col=0)
dta_atac = pd.read_csv(atac_path, index_col=0)
dta = pd.concat([dta_rna, dta_atac], axis = 1)
dta = dta.T
print(dta.shape)
```

```python
adata = data_process(dta, min_genes = 10, min_cells = 1, mt_pct = 10, 
                     npcs = 50, oversd = None)
index = adata.obs.index
index = [idx for idx, s in enumerate(index) if 'atac' in s][0]
dta_rna_use = adata.obsm['X_pca'][range(index), :]
dta_atac_use = adata.obsm['X_pca'][index:, :]
print(dta_rna_use.shape)
print(dta_atac_use.shape)
```

```python
# plot pca results
import matplotlib.pyplot as plt
import matplotlib
n_samples = dta_rna_use.shape[0]
label = np.concatenate((np.repeat(0, n_samples), np.repeat(1, n_samples)))
color = np.array(['r', 'g'])
plt.scatter(adata.obsm['X_pca'][:,0], adata.obsm['X_pca'][:, 1], c=color[label],s=1)
```

```python
n_samples = dta_rna_use.shape[0]
Wx = np.cov(dta_rna_use)
Wy = np.cov(dta_atac_use)
Wx.shape
```

```python
# manifold alignment
manifold_res = manifold_alignment(Wx = Wx, Wy = Wy, Wxy = None, mu = 0.9, d = 30, normlized = True, eps = 1e-5)
manifold_res.shape
```

```python
# plot manifold alignment results
label = np.concatenate((np.repeat(0, n_samples), np.repeat(1, n_samples)))
color = np.array(['r', 'g'])
plt.scatter(manifold_res[:,0], manifold_res[:, 1], c=color[label], s=1)
```

```python
# visualization
from sklearn.manifold import TSNE
import umap
# tSNE
manifold_embedded_tsne = TSNE(n_components=2).fit_transform(manifold_res)
print(manifold_embedded_tsne.shape)
# UMAP
reducer = umap.UMAP()
manifold_embedded_umap = reducer.fit_transform(manifold_res)
print(manifold_embedded_umap.shape)
```

```python
# print t-sne results
label = np.concatenate((np.repeat(0, n_samples), np.repeat(1, n_samples)))
color = np.array(['r', 'g'])
plt.scatter(manifold_embedded_tsne[:,0], manifold_embedded_tsne[:, 1], c=color[label], s=1)
```

```python
# plot UMAP results
plt.scatter(manifold_embedded_umap[:,0], manifold_embedded_umap[:, 1], c=color[label], s=1)
```
